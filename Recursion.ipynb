**Q.2 Minimum Operations to Convert word1 to word2 **

```python
def minDistance(word1, word2):
    m, n = len(word1), len(word2)
    dp = [[0] * (n + 1) for _ in range(m + 1)]

    for i in range(m + 1):
        for j in range(n + 1):
            if i == 0:
                dp[i][j] = j
            elif j == 0:
                dp[i][j] = i
            elif word1[i-1] == word2[j-1]:
                dp[i][j] = dp[i-1][j-1]
            else:
                dp[i][j] = 1 + min(dp[i-1][j-1], dp[i-1][j], dp[i][j-1])

    return dp[m][n]

# Test cases
print(minDistance("horse", "ros"))      # Output: 3
print(minDistance("intention", "execution"))  # Output: 5
```

How the Dynamic Programming Approach Works:

Base Case:


If either string is empty (i == 0 or j == 0), the edit distance is simply the length of the other string, as it would take that many insertions or deletions to match.

Recursive Case:

If the current characters of both strings are the same (word1[i-1] == word2[j-1]), no operation is needed, so dp[i][j] = dp[i-1][j-1].
Otherwise, choose the minimum of the three possible operations (insert, delete, replace), and add 1 to it, indicating that an operation was performed.


Dynamic Programming vs. Recursion: While the above solution uses dynamic programming (DP) to avoid recalculating the same subproblems, the recursive nature of the problem still exists conceptually. DP optimizes this by storing the results of these subproblems.



**Q.3 Find the Maximum Value in an Array**

```python
def find_max(arr):
    if len(arr) == 1:
        return arr[0]
    else:
        max_of_rest = find_max(arr[1:])
        return max(arr[0], max_of_rest)

# Test case
arr = [13, 1, -3, 22, 5]
print("Maximum value in the array is:", find_max(arr))  # Output: 22
```

**Explanation:**

Base Case: If the array contains only one element, that element is the maximum, so the function returns it directly.

Recursive Case:
If the array has more than one element, the function compares the first element (arr[0]) with the maximum value of the rest of the array (arr[1:]). The function calls itself on the rest of the array, gradually reducing the problem size until it reaches the base case.

This recursive process continues until the entire array has been reduced to individual elements, and the maximum value is found by comparing elements pairwise.

**Q.4 Find the Sum of Values in an Array**

```python
def find_sum(arr):
    if len(arr) == 0:
        return 0
    else:
        return arr[0] + find_sum(arr[1:])

# Test case
arr = [92, 23, 15, -20, 10]
print("Sum of the values in the array is:", find_sum(arr))  # Output: 120
```

**Explanation:**


**Base Case: **

If the array is empty, the sum is 0 because there are no elements to add.

**Recursive Case:**
If the array is not empty, the function adds the first element (arr[0]) to the sum of the rest of the array (arr[1:]). The function calls itself on the remainder of the array until it is reduced to an empty array, at which point the base case is reached and the total sum is returned.

This recursive approach gradually reduces the problem size by summing the first element with the sum of the remaining elements, continuing until the base case is reached.



**Q.5 Armstrong Number Detection**

```python
def is_armstrong_number(num):
    digits = str(num)
    n = len(digits)
    sum_of_powers = sum(int(digit) ** n for digit in digits)
    return sum_of_powers == num

# Test cases
num1 = 153
num2 = 134

if is_armstrong_number(num1):
    print(f"{num1} is an Armstrong number: Yes")
else:
    print(f"{num1} is an Armstrong number: No")

if is_armstrong_number(num2):
    print(f"{num2} is an Armstrong number: Yes")
else:
    print(f"{num2} is an Armstrong number: No")

```

Base Case:

The recursion isnt explicit in this case. Instead, it's handled by Pythons internal sum() function that applies the recursive operation of summing up each digit raised to the power of n.

Recursive Case:

The sum() function internally iterates over each digit and applies the power operation recursively to accumulate the total sum, which is then compared to the original number.
